# **2.How does broadcasting work in NumPy ?**

-  Broadcasting in NumPy is a mechanism that allows NumPy to perform element-wise operations on arrays of different shapes without explicitly copying data.

**How broadcasting works:**

-NumPy compares the shapes of two arrays from right to left and follows these rules:

1.The dimensions are compatible if

- they are equal, or

- one of them is 1.

- 2.If the rules are satisfied, NumPy stretches the smaller array (virtually, not in memory) to match the larger array.

- 3.The operation is then performed element-wise.

**Example 1:**

    Array + Scalar
    import numpy as np

    a = np.array([1, 2, 3])
    b = 5

    print(a + b)


**Output:**

     [6 7 8]


- Here, the scalar 5 is broadcast to [5, 5, 5].

**Example 2:**
    2D array + 1D array
    a = np.array([[1, 2, 3],
              [4, 5, 6]])

    b = np.array([10, 20, 30])

    print(a + b)


**Output:**

    [[11 22 33]
    [14 25 36]]


**The 1D array is broadcast across each row.**

**When broadcasting fails:**

    a = np.array([[1, 2],
              [3, 4]])

    b = np.array([1, 2, 3])

# a + b → Error (shapes not compatible)

#Key point:

**Broadcasting makes NumPy fast and memory-efficient because it avoids creating large temporary arrays.**

**Broadcasting lets NumPy handle operations between arrays of different shapes automatically, as long as their dimensions are compatible**

# **17. Explain the difference between apply() and map() in Pandas.**


 **1. map()**

- Used mainly with Series.

- Applies a function element-wise to each value in the Series.

- Can take a function, dictionary, or mapping.

**Example:**

    import pandas as pd

    s = pd.Series([1, 2, 3, 4])
    # Square each element using map
    squared = s.map(lambda x: x**2)
    print(squared)


    Output:

    0     1
    1     4
    2     9
    3    16
    |dtype: int64

**2. apply()**

- Can be used with Series or DataFrame.

- Applies a function along rows or columns in a DataFrame, or element-wise in a Series.

- More flexible than map() for complex operations.

**Example with Series:**

    s = pd.Series([1, 2, 3, 4])
    # Add 10 to each element using apply
    added = s.apply(lambda x: x + 10)
    print(added)


    Output:

    0    11
    1    12
    2    13
    3    14
    dtype: int64


**Example with DataFrame:**

    df = pd.DataFrame({'A':[1,2], 'B':[3,4]})
    # Sum across rows

    row_sum = df.apply(lambda x: x.sum(), axis=1)
    print(row_sum)


    Output:

    0    4
    1    6
    dtype: int64

**Key Difference:**

- map() → element-wise, mainly for Series, simpler functions.

- apply() → element-wise or row/column-wise, works for Series and DataFrames, more flexible.

# **21.Why is NumPy’s array slicing faster than Python’s list slicing?**


**NumPy’s array slicing is faster than Python’s list slicing because:**

- Contiguous Memory Storage – NumPy arrays store data in continuous memory blocks, while Python lists store references to objects scattered in memory.

- Vectorized Operations – Slicing returns a view on the same data (not a copy), allowing fast access and computation.

- Low-level Implementation – NumPy is implemented in C, so operations like slicing are highly optimized compared to Python’s interpreted list operations.

**In short: NumPy slicing is memory-efficient and computationally faster than Python list slicing.**
